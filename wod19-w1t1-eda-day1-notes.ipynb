Roadmap:
1. Exploratory data analysis - informal way to look at data and see what is interesting in it
2. Descriptive statistics - how to summarize data
3. Estimation - get an idea for how big a (set of) parameter is (are)
4. Hypothesis testing - have a question and want a yes or no
5. Design of experiments - for A/B testing, if I want to answer a certain question and want to gather data, how should I do this, how many data points do I need, how should I sample my data

Will not touch upon:
* Regression
* Machine learning - engineering perspective, gives you an output: yes or no, or predictions
* Bayesian data analysis

```python
import numpy as np
n = 20
labels = ['age']+['%02d'%i for i in range(1, n+1)]
l = [100] + [x for x in np.random.rand(n)*10+20]
with open('heights_sequoias1.csv', 'w') as fout:
    # print >> fout, ','.join(str(x) for x in labels) #python 2
    print(','.join(str(x) for x in labels), file = fout)
    print(','.join(str(x) for x in l), file = fout)
    for age in range(200, 2901, 100):
        newl = [age] + [4*np.random.rand(1)[0]+x for x in l[1:]]
        print(','.join(str(l) for l in newl), file=fout)
        l = newl
```
